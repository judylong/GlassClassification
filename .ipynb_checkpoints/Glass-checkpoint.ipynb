# Glass Classification

## Task
This is a multiclassification problem to classify glass into one of 6 types based on refraction index and percent composition of component elements.

### Notes on dataset:
- 214 samples
- 9 input features (one being refraction index (RI) and the rest are percent composition of other elements)
- there are no missing values for any of the input features
- the output is a 'Type' with class values labeled 1, 2, 3, 5, 6, 7
- there are no samples in the dataset with class value '4'
- there are somehow 26 samples whose composite metal percentages add up to over 100% (abeit slightly - max: 100.10)

#### Technical Highlights
- Explores 2 types of oversampling methods: random and SMOTE to deal with unbalanced classes
- Uses nested gridsearch to find an unbiased/less biased performance measure
- Uses randomized gridsearch search larger parameter grids to save computational effort/time
- Imports module from imblearn for SMOTE

# Model Selection

```python
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.metrics import cohen_kappa_score, make_scorer

X = glass.loc[:, :'Fe']
y = glass['Type']

kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=3)
kappa = make_scorer(cohen_kappa_score)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import Perceptron
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier

#pipelines for distance based algorithms
pipeline_svc = Pipeline([('scaling', StandardScaler()), ('clf', SVC())])
pipeline_svc_balanced = Pipeline([('scaling', StandardScaler()), ('clf', SVC(class_weight='balanced'))])
pipeline_knn = Pipeline([('scaling', StandardScaler()), ('clf', KNeighborsClassifier())])

pipes_unsampled = {'svc' : pipeline_svc_balanced, 
         'knn': pipeline_knn, 
         'perceptron': Perceptron(class_weight='balanced'),
         'nb': GaussianNB(), 
         'dtree': DecisionTreeClassifier(), 
         'rforest': RandomForestClassifier(), 
         'boostedTree': GradientBoostingClassifier()
        }

pipes = {'svc' : pipeline_svc, 
         'knn': pipeline_knn, 
         'perceptron': Perceptron(),
         'nb': GaussianNB(), 
         'dtree': DecisionTreeClassifier(), 
         'rforest': RandomForestClassifier(), 
         'boostedTree': GradientBoostingClassifier()
        }

param_grids = {
    'svc' : {'clf__C': [.001, .01, .1, 1], 'clf__gamma': [1, 2, 3, 4]}, 
    'knn': {'clf__n_neighbors': [1,2,3,4,5]}, 
    'perceptron': {'penalty':['l2', 'l1', 'elasticnet'], 'alpha': [.00001, .0001, .001]}, 
    'nb': {}, 
    'dtree': {'max_depth': [1,2,3,4,5]}, 
    'rforest': {'n_estimators': [5, 10, 20, 30]}, 
    'boostedTree': {'n_estimators': [50, 100, 150]}
}
```

```python
def add_mean_std_to_df(df):
    df['mean'] = df.mean(axis=1)
    df['std'] = df.std(axis=1)
    return df
```

# Imbalanced Classes

### Original Data with out-of-box classifiers
Here, we check the baseline performances of several algorithms out-of-the-box without changing the imbalanced classes. As noted during the data exploration above, this is a multi-class classification problem with imbalanced classes. Therefore, Cohens Kappa statistic has been chosen as our performance metric.

```python
def no_sampling_cv(pipelines, X, y, cv, scorer):
    unbiased_scores = {}
    
    for train_idx, test_idx in cv.split(X, y):
        X_train_fold, y_train_fold = X.iloc[train_idx], y.iloc[train_idx]
        X_test_fold, y_test_fold = X.iloc[test_idx], y.iloc[test_idx]
        
        for name, pipe in pipelines.items():
            estimator = GridSearchCV(pipe, param_grid=param_grids[name], scoring=scorer, cv=cv)
            estimator.fit(X_train_fold, y_train_fold)
            # calculate unbiased score on outer cv
            score = cohen_kappa_score(y_test_fold, estimator.best_estimator_.predict(X_test_fold))
#             print(name, score)
            try:
                unbiased_scores[name].append(score)
            except KeyError:
                unbiased_scores[name] = [score]

    return unbiased_scores
```

```python
scores = no_sampling_cv(pipes_unsampled, X, y, kfold, kappa)
```

```python
scores_df = pd.DataFrame.from_dict(scores, orient='index')
add_mean_std_to_df(scores_df)
scores_df['mean'].sort_values()
```

# Using oversampling to deal with imbalanced classes
There are a number of approaches to handle data with imbalanced classes, such as undersampling and oversampling. 
The class with the least number of samples is Type 6 with 9 samples and the class with the most is Type 2 with 76 samples, which is over 8 times more. Undersampling would result in creating a classifier that is trained with a lot fewer samples. Therefore, I've chosen to explore 2 oversampling methods: random oversampling and SMOTE (Synthetic Minority Over-sampling Technique). Oversampling is executed on the training samples within each fold of cross validation, ensuring that the model being built never sees the test data (data leakage).

## Note on implementation:
Nested cross validation is used to obtain an unbiased/less biased performance measure for model selection.

# Balancing Classes with Random Oversampling
This implementation of oversampling will fill in the difference in sample sizes between each minority class and the majority class by drawing randomly with replacement.

```python
from sklearn.utils import resample

def oversample(X, y, label, random_state=123):
    df = pd.concat([X, y], axis=1)
    max_samples = df[label].value_counts().max()
    label_values = df[label].unique()
    df_by_label = dict(list(df.groupby(label)))
    
    oversampled = []
    
    for val in label_values:
        if df_by_label[val].shape[0] == max_samples:
            oversampled.append(df_by_label[val])
        else:
            original_df = df_by_label[val]
            num_to_draw = max_samples - original_df.shape[0]
            resampled_df = resample(df_by_label[val], replace=True, n_samples=num_to_draw, random_state=random_state)
            oversampled += [original_df, resampled_df]
            
    oversampled_df = pd.concat(oversampled)
    return oversampled_df.drop(label, axis=1), oversampled_df[label]
```

```python
def random_oversampled_cv(pipelines, X, y, cv, scorer):
    unbiased_scores = {}
    
    for train_idx, test_idx in cv.split(X, y):
        X_oversampled, y_oversampled = oversample(X.iloc[train_idx], y.iloc[train_idx], 'Type')
        X_test_fold, y_test_fold = X.iloc[test_idx], y.iloc[test_idx]
    
        fold_test_scores = {}
        fold_train_scores = {}

        for name, pipe in pipelines.items():
            estimator = GridSearchCV(pipe, param_grid=param_grids[name], scoring=scorer, cv=cv)
            estimator.fit(X_oversampled, y_oversampled)
            
            # calculate unbiased score on outer cv
            score = cohen_kappa_score(y_test_fold, estimator.best_estimator_.predict(X_test_fold))
#             print(name, score)
            try:
                unbiased_scores[name].append(score)
            except KeyError:
                unbiased_scores[name] = [score]

    return unbiased_scores
```

```python
random_unbiased_scores = random_oversampled_cv(pipes, X, y, kfold, kappa)
```

```python
random_unbiased_scores_df = pd.DataFrame(random_unbiased_scores).T
add_mean_std_to_df(random_unbiased_scores_df)
random_unbiased_scores_df['mean'].sort_values()
```

# Balancing Classes with Oversampling (SMOTE)

```python
from imblearn.over_sampling import SMOTE

def smote_oversampled_cv(pipelines, X, y, cv, scorer):
    unbiased_scores = {}
    
    for train_idx, test_idx in cv.split(X, y):
        X_fold, y_fold = X.iloc[train_idx], y.iloc[train_idx]

        X_resampled, y_resampled = SMOTE().fit_sample(X_fold, y_fold)
        X_test_fold, y_test_fold = X.iloc[test_idx], y.iloc[test_idx]

        fold_test_scores = {}
        fold_train_scores = {}

        
        for name, pipe in pipelines.items():
            pipe.fit(X_resampled, y_resampled)
            
            estimator = GridSearchCV(pipe, param_grid=param_grids[name], scoring=scorer)
            estimator.fit(X_resampled, y_resampled)
            # calculate unbiased score on outer cv
            score = cohen_kappa_score(y_test_fold, estimator.best_estimator_.predict(X_test_fold))
#             print(name, score, estimator.best_params_)
            try:
                unbiased_scores[name].append(score)
            except KeyError:
                unbiased_scores[name] = [score]

    return unbiased_scores
```

```python
smote_unbiased_scores = smote_oversampled_cv(pipes, X, y, kfold, kappa)
```

```python
smote_unbiased_scores_df = pd.DataFrame(smote_unbiased_scores).T
add_mean_std_to_df(smote_unbiased_scores_df)
smote_unbiased_scores_df['mean'].sort_values()
```

The above experiments show that tree ensembles (random forests and gradient boosed trees) work best with this dataset, regardless of oversampling technique used (or lackthereof). Even as the best models, they do not yield a performce score above ~.7. The gains from using oversampling are only around .01, but this may only be due to this particular combination of initial random state for the splits and the initial random state for the tree-based models. If allowed to be completely random on each initialization, there would be no distinguishable/significant gain from the use of oversampling methods.

# Final Model

A final cross validation with finer hyperparameter tuning to fit the final model without oversampling. 

## Random Forest

```python
rforest = RandomForestClassifier(random_state=10)

N_ESTIMATORS = range(1, 100, 5)
param_grid = dict(n_estimators=N_ESTIMATORS)
estimator = GridSearchCV(rforest, param_grid=param_grid, scoring=kappa, cv=kfold)
estimator.fit(X, y)
print(estimator.best_params_)
print(estimator.best_score_)
```

```python
x = N_ESTIMATORS
y_train = estimator.cv_results_['mean_train_score']
y_test = estimator.cv_results_['mean_test_score']
plt.plot(x, y_train, label='train score')
plt.plot(x, y_test, label='test score')
plt.legend()
plt.xlabel('number of estimators')
plt.ylabel('score')
plt.title('Score vs. Number of Estimators - Random Forest')
plt.figure(figsize=(6,8))
plt.show()
```

## Gradient Boosted Trees

```python
from sklearn.model_selection import RandomizedSearchCV
boostedTree = GradientBoostingClassifier()

LEARNING_RATE_RANGE = np.linspace(.001, 1, 25)
MAX_DEPTH_RANGE = np.arange(2, 15)

param_distributions = dict(learning_rate=LEARNING_RATE_RANGE, max_depth=MAX_DEPTH_RANGE)
estimator = RandomizedSearchCV(boostedTree, param_distributions=param_distributions, scoring=kappa, cv=kfold)
estimator.fit(X,y)
print(estimator.best_params_)
print(estimator.best_score_)
```

From the above experiments, we can see that Random Forest with n_estimators=31 works best.
