# Gradient Tree Boosting

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.cross_validation import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
gboost = GradientBoostingClassifier(n_estimators=100)

gboost.fit(X_train, y_train)
```

```python
n_estimators = 100
test_score = np.zeros((n_estimators,), dtype=np.float64)
for i, y_pred in enumerate(gboost.staged_predict_proba(X_test)):
#     print(y_pred) try labelencoding or label+one hot encoding
    test_score[i] = gboost.loss_(y_test, y_pred)

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.title('Deviance')
plt.plot(np.arange(n_estimators) + 1, gboost.train_score_, 'b-',
         label='Training Set Deviance')
plt.plot(np.arange(n_estimators) + 1, test_score, 'r-',
         label='Test Set Deviance')
plt.legend(loc='upper right')
plt.xlabel('Boosting Iterations')
plt.ylabel('Deviance')
plt.show() #devuance doesn't max out at 100--this is a funky chart..not trusting
```

```python
def oversample(df, label, random_state=123):
    max_samples = df[label].value_counts().max()
    label_values = df[label].unique()
    df_by_label = dict(list(df.groupby(label)))
    
    oversampled = []
    
    for val in label_values:
#         print(val, df_by_label[val].shape[0])
        if df_by_label[val].shape[0] == max_samples:
            oversampled.append(df_by_label[val])
        else:
            resampled_df = resample(df_by_label[val], replace=True, n_samples=max_samples, random_state=random_state)
            oversampled.append(resampled_df)
    return pd.concat(oversampled)
```
